```python
import pandas as pd
from PIL import Image
import os
import io
```

```python


# Folder where your .parquet files are stored
parquet_folder = "exports\\nature"  # adjust if needed
output_folder = "I:\dataset\\nature128\\nature128_images"
os.makedirs(output_folder, exist_ok=True)

# Get all .parquet files in the folder
parquet_files = [f for f in os.listdir(parquet_folder) if f.endswith(".parquet")]
```

```python
img_count = 0
for file in parquet_files:
    path = os.path.join(parquet_folder, file)
    print(f"Reading: {path}")
    
    # Load Parquet file
    df = pd.read_parquet(path)
    
    for idx, row in df.iterrows():
        image_dict = row["image"]
        image_bytes = image_dict["bytes"]
        
        # Convert bytes to PIL Image
        img = Image.open(io.BytesIO(image_bytes)).convert("RGB")
        
        # Save image
        img_path = os.path.join(output_folder, f"image_{img_count:04d}.png")
        img.save(img_path)
        print(f"Saved: {img_path}")
        img_count += 1
```
